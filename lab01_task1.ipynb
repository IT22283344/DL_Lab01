```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
exp_data = np.random.exponential(scale=1.0, size=(100000,))
uniform_data = np.random.uniform(low=0.0, high=2.0, size=(100000,))
normal_data = np.random.normal(loc=1.0, scale=0.5, size=(100000,))
```

```python

# Plot histograms
plt.figure(figsize=(12, 6))
plt.hist(exp_data, density=True, bins=100, histtype="step", color="blue", label="Exponential")
plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc="upper left")
plt.title("Exponential")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
plt.hist(uniform_data, density=True, bins=100, histtype="step", color="green", label="Uniform")
plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc="upper left")
plt.title("Uniform Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
plt.hist(normal_data, density=True, bins=100, histtype="step", color="red", label="Normal")
plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc="upper left")
plt.title("Normal Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2
```

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title('3D Plot of Z = X + Y')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import requests
from io import StringIO

# Load CSV from URL using headers to bypass 403 error
url = "https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv"
headers = {'User-Agent': 'Mozilla/5.0'}
response = requests.get(url, headers=headers)

# Check if successful
if response.status_code == 200:
    df = pd.read_csv(StringIO(response.text))
else:
    raise Exception(f"Failed to fetch data: {response.status_code}")

# Select relevant numerical features
stats = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
df_stats = df[stats]
```

```python
# Calculate correlations
pearson_corr = df_stats.corr(method='pearson')
spearman_corr = df_stats.corr(method='spearman')
```

```python
# Set plot style
sns.set(style="whitegrid")

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Set plot style
sns.set(style="whitegrid")

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Pearson heatmap
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', fmt=".2f", ax=axes[0])
axes[0].set_title(" Pearson Correlation")

# Spearman heatmap
sns.heatmap(spearman_corr, annot=True, cmap='viridis', fmt=".2f", ax=axes[1])
axes[1].set_title(" Spearman Rank Correlation")

plt.tight_layout()
plt.show()
```
